Code which changes the label data from its statistical
distribution to the corresponding labels given. The resulting dataframe is then written to a csv file for data visualization.


```
import pandas as pd
```

```
df = pd.read_csv("drug_consumption.data", header=None)
```

```
df.head()
df = df.drop(df.columns[0],axis=1)
df.columns = ["Age", "Gender", "Education", "Country", "Ethnicity","Nscore",
              "Escore","Oscore","Ascore","Cscore","Imp","SS","Alc","Amph",
              "Amyl","Benz","Caff","Cann","Choc","Coke","Crack","Ect",
              "Heroin","Keta","Legalh","LSD","Meth","Shrooms","Nico","Semer","VSA"]
df.head()
```

```
##Switch the distribution statistics to corresponding values for visualizations.
#Switch all real values for ages to the corresponding age ranges.
df.loc[(df.Age == -0.9519700000000001), 'Age'] = "18-24"
df.loc[(df.Age == -0.07854), 'Age'] = "25-34"
df.loc[(df.Age == 0.49788000000000004), 'Age'] = "35-44"
df.loc[(df.Age == 1.09449), 'Age'] = "45-54"
df.loc[(df.Age == 1.82213), 'Age'] = "55-64"
df.loc[(df.Age == 2.59171), 'Age'] = "65+"

#Switch gender statistics to either male or female, dependent on participant.
df.loc[(df.Gender == 0.48246000000000006), 'Gender'] = "Female"
df.loc[(df.Gender == -0.48246000000000006), 'Gender'] = "Male"

#Switch education from its statistical distribution into the corresponding levels.
df.loc[(df.Education == -0.059210000000000006), 'Education'] = "Diploma" #Professional Diploma
df.loc[(df.Education == 1.98437), 'Education'] = "PhD" #PostDoctorate
df.loc[(df.Education == 1.16365), 'Education'] = "MA" #Masters
df.loc[(df.Education == -1.22751), 'Education'] = "L18" #Left school at 18
df.loc[(df.Education == -1.7379), 'Education'] = "L16" #Left school at 16
df.loc[(df.Education == 0.45468000000000003), 'Education'] = "UG" #Undergraduate degree
df.loc[(df.Education == -0.6111300000000001), 'Education'] = "Incomplete" #Incomplete undergraduate degree
df.loc[(df.Education == -2.43591), 'Education'] = "LB16" #Left school before 16
df.loc[(df.Education == -1.43719), 'Education'] = "L17" #Left school at 17

#Switch country distribution to country names.
df.loc[(df.Country == 0.9608200000000001), 'Country'] = "UK"
df.loc[(df.Country == 0.24923), 'Country'] = "CA"
df.loc[(df.Country == -0.57009), 'Country'] = "USA"
df.loc[(df.Country == -0.28519), 'Country'] = "Other"
df.loc[(df.Country == -0.09765), 'Country'] = "AUS"
df.loc[(df.Country == 0.21128000000000002), 'Country'] = "IRL"
df.loc[(df.Country == -0.46841000000000005), 'Country'] = "NZ"

#Dropped ethnicity as 92% of participants are white and the remaining ethnicities were
#only composed of a few participants each.
df = df.drop(columns = ['Ethnicity'],axis=1)
df.head()
```

```
df.to_csv("drug_cons_data.csv",index=False)
```
